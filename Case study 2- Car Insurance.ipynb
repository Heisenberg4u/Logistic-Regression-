# Let's build our model

```python
#First thing is to get over data ready
```

```python
insurance.head()
```

```python
#^O/P=The data has both the numerical values as well as character data,before performing any algorithm we need to take care of
#Character data as we can't perform any maths on it so lets separate numerical columns & char. columns & then for Char. columns
#we will create dummy models.
```

```python
numerical_cols = insurance.select_dtypes(include=[np.number])
```

```python
numerical_cols 
```

```python
numerical_cols.head()  
```

```python
#^O/P=it only has num. columns & similarly we can separate the categorical data or char. data
```

```python
character_cols = insurance.select_dtypes(include=['object'])
```

```python
character_cols
```

```python
character_cols.drop(columns=['CallStart','CallEnd'],inplace=True)
```

```python
character_dummies = pd.get_dummies(character_cols)
```

```python
character_dummies 
```

```python
character_cols = insurance.select_dtypes(include=['object'])       # again run it to check
```

```python
character_cols 
```

```python
character_dummies = pd.get_dummies(character_cols)                   # again run it to check 
```

```python
character_dummies 
```

```python
character_dummies.columns
```

```python
insurance_combined = pd.concat([numerical_cols,character_dummies],axis=1)
```

```python
insurance_combined      #o/p=now we have all numerical values
```

```python
insurance_combined.columns  #o/p=CarInsurance is our Y & everything else is our X
```

```python
# dependent variable
Y = insurance_combined['CarInsurance']
```

```python
# independent variables
X = insurance_combined.drop(columns='CarInsurance')
```

```python
#Now next is to divide the data in Train & Test data so for that
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,train_size=0.7,random_state=100)
```

```python
X_train.shape,Y_train.shape,X_test.shape,Y_test.shape
```

# Now Let's build our model

```python
X_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(random_state=100)
```

```python
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
Y_pred 
```

# Let's check performance of the Model

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(Y_test, Y_pred)
```

```python
accuracy_score(Y_test, Y_pred)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
#Soft prediction
area_under_roc = roc_auc_score(Y_test,model.predict_proba(X_test)[:,1])
```

```python
area_under_roc 
```

```python
from sklearn.metrics import roc_curve
```

```python
fpr,tpr,thresholds = roc_curve(Y_test,model.predict_proba(X_test)[:,1])
```

```python
plt.plot(fpr,tpr)
```

```python
#And now we can check the different thresholds at these edge points and use those values for thresholds & see what is the 
#accuracy based on hard prediction
```

```python
#If our accuracy is 68% then from this curve we can see our accuracy can't go beyond 75%.Had there been sharp edges on left & 
#Top side then we would have used those values on the edges & see there is a significant jump in the accuracy but there will be
#few % jump in the accuracy bcz the curve is very smooth,there are no sharp edges
```

```python
#Overall we can say it is a decent model with a close to 70% accuracy but definetely we should try other models.We need to try
#Different EDA techniques or different thresholds or we need to build another models maybe Random Forest to see if the data is
#explained better by that model or we can also change the hyperparameters & see if we get better accuracy
```
